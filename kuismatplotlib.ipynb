```python
import matplotlib.pyplot as plt
import numpy as np
import mpl_toolkits.axisartist as axisartist

def setup_axes(fig, rect):
    ax = axisartist.Subplot(fig, rect)
    fig.add_subplot(ax)

    return ax

fig = plt.figure(figsize=(12, 2))

x = np.arange(0,100)

ax1 = setup_axes(fig, "121")
ax1.plot(x*2, marker='*')
ax1.set_xlabel("x")
ax1.set_ylabel("y")

ax2 = setup_axes(fig, "122")
ax2.plot(x**2,  marker='*')
ax2.set_xlabel("x")
ax2.set_ylabel("z")

plt.show()
```
